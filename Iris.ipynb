```python

colors = {"Iris-virginica":"purple", "Iris-setosa": "blue", "Iris-versicolor":"green"}

def showDistributions(feature1):
    plt.figure(figsize=(30,30))
    plt.subplot(6,6,1)
    plt.scatter(flowers_df[feature1], flowers_df['species'], c=flowers_df['species'].map(colors))
    plt.title("{} distribution".format(feature1))
    plt.xlabel(feature1)
    plt.ylabel('species')
    plt.show()
    
showDistributions('sepal_length')
showDistributions('petal_length')
showDistributions('sepal_width')
showDistributions('petal_width')
```

```python
from sklearn.cluster import KMeans
```

```python
flowers_df
```

```python
x = flowers_df.drop(['species'], axis = 1)
y = flowers_df['species']
```

```python
print('features')
print(x)
print('*****' * 10)
print ('targets')
print(y)
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 500, n_init =10, random_state =0 )
model= kmeans.fit_predict(x)
```

```python
kmeans.cluster_centers_
```

```python
species = {'Iris-setosa': 0, 'Iris-virginica' : 1, 'Iris-versicolor': 2}
irisdata = flowers_df.copy()

irisdata['species'] = irisdata['species'].map(species)
irisdata['predicted'] = model
```
